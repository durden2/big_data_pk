```python
import pandas as pd
from sklearn.ensemble import ExtraTreesClassifier
```

```python
data = pd.read_csv('diabetes.csv')
col_names = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

for col_name in col_names:
    average = round(data[col_name].mean(skipna=True), 1 if col_name == 'BMI' else 0)
    data[col_name] = data[col_name].replace(0, average)

data['DiabetesPedigreeFunction'] = data['DiabetesPedigreeFunction'].round(3)

data.to_csv('diabetes_mean.csv', index=None)
```
